```python
import pandas as pd
df1 = pd.read_csv('Cond4n_e104.csv')
df1.head()
```

```python
df2 = pd.read_csv('COND2n.csv')
df2.head()
```

データ２の列をデータ１の列名に合わせます。
![image.png](https://qiita-image-store.s3.ap-northeast-1.amazonaws.com/0/223763/7540f380-1ff1-34d9-7175-17bd066f9c87.png)


```python
#データ２の列をデータ１の列名に合わせます。
df2_1=df2.rename(columns={'Time': 'UP_TIME', 'Power': 'POWER', 'Temperature': 'TEMP', 'Outcome': 'ERR_CD'})\
    .drop(['Pressure','Uptime','Status'],axis=1)
df2_1
```

M_CDに当たる列がデータ２のCOND2n.csvには存在しないのでNULLが入ります。
![image.png](https://qiita-image-store.s3.ap-northeast-1.amazonaws.com/0/223763/c9334cfe-0535-741b-d586-56b471e64fd5.png)

```python
#appendを使う方法
df1.append(df2_1)
```

```python
#concatを使う方法
pd.concat([df1,df2_1])
```
