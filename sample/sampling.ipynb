```python
#層化サンプリング
Stratified_df=df.groupby('M_CLASS', group_keys=False)\
    .apply(lambda x: x.sample(frac=0.002, random_state = 1))
```

scikit-learnのStratifiedShuffleSplitを使ってテストデータを作る方法<BR>
https://www.randomlyforest.com/entry/2019/01/14/215102
<BR>sklearn.model_selection.StratifiedShuffleSplit — scikit-learn 0.23.1 documentation<BR>
https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.StratifiedShuffleSplit.html

```python
#層化サンプリング StratifiedShuffleSplit利用パターン
from sklearn.model_selection import StratifiedShuffleSplit
sample = StratifiedShuffleSplit(n_splits = 1,train_size = 0.002,test_size = 0.01, random_state = 1)
for train_,test_ in sample.split(df, df['M_CLASS']):
    StratifiedShuffleSplit_df = df.loc[train_]
#    chunk_test = df1.loc[test_]
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
fig, axes = plt.subplots(nrows=1, ncols=4,figsize=(10.0, 5.0))
#plt.title(titles[i], fontsize=20)
#元データ
df[['M_CLASS']].groupby(['M_CLASS']).size().plot(ax=axes[0],  kind='bar' ,title='original')
#層化サンプリング
Stratified_df[['M_CLASS']].groupby(['M_CLASS']).size().plot(ax=axes[1], kind='bar',title='Stratified_df')
#層化サンプリング StratifiedShuffleSplit
StratifiedShuffleSplit_df[['M_CLASS']].groupby(['M_CLASS']).size().plot(ax=axes[2], kind='bar',title='StratifiedShuffleSplit')
#ただのランダムサンプリング
df[['M_CLASS']].sample(frac=0.002,random_state=1).groupby(['M_CLASS']).size().plot(ax=axes[3],  kind='bar',title='random')

```
